### Motivation + Data Source
If you know me well, you'd know that I really enjoy working with language data. ðŸ˜Š<p> 

We talk in many parts of speech and <b>adjectives</b> are one of them. Technically, adjectives are used to modify nouns, but less technically, we use them to <i>describe</i> and add <i>sentiment</i> to the topics we might be speaking about.<p> 
    
The polarity of an adjective on its own is quite obvious.<p> For instance:
- respectable &rarr; positive
- harsh &rarr; negative
    
But what if we added conjunctions between adjectives? How does that change the meaning of the entire phrase?<p>
For instance:
- respectable <b>or</b> harsh
- respectable <b>and</b> harsh
- respectable <b>but</b> harsh
- respectable <b>yet</b> harsh

...and what does this look like in the context of movie reviews?

The dataset that I am using for my analysis contains over 100,000 coordinated adjective tokens from user-supplied movie reviews from IMDB. The dataset was provided to me by my professor at Columbia University, but you can also obtain them from my repository or from [here](https://web.stanford.edu/class/cs424p/assignments/cs424p-assign01.html).

<b>What do each of these columns name mean? </b>

- <b>PolMatch</b> &rarr; 1 if the cols. 3 and 8 values match, else 0.<br>
- <b>Adj1</b> &rarr; The left adj in the coordination.<br>
- <b>Adj1Pol</b> &rarr; The polarity of Adj1 according to the Harvard Inquirer.<br>
- <b>Adj1Prefix, Adj1Stem, Adj1Suffix</b> &rarr; A heuristic parse of Adj1. For example, disinterested â†’ (dis, interest, ed) and unhappy â†’ (un, happy,)<br>
- <b>Adj2</b> &rarr; The right adj in the coordination.<br>
- <b>Adj2Pol</b> &rarr; The polarity of Adj2 according to the Harvard Inquirer.<br>
- <b>Adj2Prefix, Adj2Stem, Adj2Suffix</b> &rarr; A heuristic parse of Adj2.<br>
- <b>BaseMatch</b> &rarr; 1 if col. 5 matches col. 10, else 0<br>
- <b>Coord</b>&rarr;The coordinator: and, but, nor, or, versus, or yet.<br>
- <b>Conj1Neg</b> &rarr; 1 if the first conjunct was negated with not or n't, else 0. (It is generally hard to tell whether such negation scopes over the entire conjunction or just the first conjunct.)<br>
- <b>Conj2Neg</b> &rarr; 1 if the second conjunct was negated with not, else 0.<br>
- <b>ClauseType</b> &rarr; The clause type of the containing sentence: DEC(larative); INT(errogative); EXC(clamative), based on the final punctuation tag.<br>
- <b>Genre</b> &rarr; The genre of the movie under review.<br>
- <b>Year</b> &rarr; The release year of the movie under review.<br>
- <b>Rating</b> &rarr; The rating of the review containing this coordination: 1...10.<br>
- <b>Helpful</b> &rarr;	The percentage of registered users who found the review containing this coordination helpful [0,1]<br>

NOTE: Descriptions retrieved from [here](https://web.stanford.edu/class/cs424p/assignments/cs424p-assign01.html).

### Preliminary Exploration of Data

```r
ggplot(data = imdb_adj_pol) + 
  geom_bar(mapping = aes(x = Coord))
```

From the bar chart above, we can already see that users tend to use the word "and." For a possible classification task, we can also assume that a bias towards the coordinator "and" because of its larger sample size compared to other coordinators.

```r
ggplot(data = imdb_adj_pol) + 
  geom_bar(mapping = aes(x = Genre, fill = ClauseType)) +
theme(axis.text.x = element_text(angle = 90, hjust = 1))
```

It looks like declaratives are most commonly used in movie reviews of all types, as indicated by the visualization above. Declaratives often indicate credibility and confidence in one's statement, so we're looking at pretty reliable dataset here.

```r
# let's examine words in which the polarity is matched
but_match <- filter(imdb_adj_pol, PolMatch == 1, Coord == "but", nchar(Genre) > 2)
```

```r
but_genre_ratings <- count(but_match, c("Genre", "Rating"))
```

```r
ggplot(data = but_genre_ratings) + 
  geom_point(mapping = aes(x = Rating, y = freq, color = Genre)) +
theme(axis.text.x = element_text(angle = 90, hjust = 1)) 
```
